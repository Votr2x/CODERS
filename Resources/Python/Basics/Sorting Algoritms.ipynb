# Sorting Algoritms

## Bubble sort

Bubble sort is a simple sorting algorithm that repeatedly steps through the list. In this algorithm, adjacent elements are compared and swapped if they are in the wrong order. The pass through the list is repeated until the list is sorted.

- Time complexity: $O(n^2)$ in the worst and average cases, $O(n)$ in the best case (when the input array is already sorted)
- Space complexity: $O(1)$

## Selection sort

The selection sort algorithm starts by finding the smallest number in the list and exchanging it with the first number. The next smallest number is found and exchanged with the second number, and so on. 

- Time complexity: $O(n^2)$ in all cases (worst, average, and best)
- Space complexity: $O(1)$

## Insertion sort

## Merge Sort Algorithm
Merge Sort is one of the most popular sorting algorithms that is based on the principle of **Divide and Conquer Algorithm.**

Here, a problem is divided into multiple sub-problems. Each sub-problem is solved individually. Finally, sub-problems are combined to form the final solution.

- Time Complexity: Merge Sort has a consistent time complexity of O(n log n) in all casesworst, average, and best. This efficiency makes it suitable for sorting large datasets. The O(n log n) time complexity arises from the recursive splitting of the array and the merging of sorted halves.

- Space Complexity: Merge Sort typically has a space complexity of O(n) due to the auxiliary arrays used during the merge process. This additional space is required to store the temporary merged subarrays.

**Algorithm:**

1. Divide:
- Split the unsorted list into two halves recursively until each sublist contains one element. This is the base case of the recursion.
2. Conquer:
- Sort each sublist recursively. This step continues until all sublists are sorted.
3. Combine (Merge):
- Merge the sorted sublists to produce the final sorted array.
- During merging, elements from both sublists are compared and placed in order in a temporary array, which becomes the sorted sublist.

#### Step-by-Step Explanation of Merge Sort

**Divide: Splitting the Array**

- **Initial Array**: [6, 5, 12, 10, 9, 1]
- Divide the array into two halves:
  - Left half: [6, 5, 12]
  - Right half: [10, 9, 1]

**Recursive Sorting**

- **Left Half: [6, 5, 12]**
  - Divide further:
    - Left: [6, 5]
    - Right: [12]
  - Continue dividing until each sublist contains only one element:
    - [6], [5], [12]
  - Merge sorted sublists:
    - Merge [6] and [5]  [5, 6]
  - Left half sorted: [5, 6, 12]

- **Right Half: [10, 9, 1]**
  - Divide further:
    - Left: [10]
    - Right: [9, 1]
  - Continue dividing until each sublist contains only one element:
    - [9], [1]
  - Merge sorted sublists:
    - Merge [9] and [1]  [1, 9]
  - Right half sorted: [1, 9, 10]

**Combine (Merge): Merge the Sorted Halves**

- Merge the two sorted halves [5, 6, 12] and [1, 9, 10]:
  - Compare elements from both lists and place them in order:
    - [1, 5, 6, 9, 10, 12]

### Final Sorted Array
After completing the merge step, the final sorted array is [1, 5, 6, 9, 10, 12].

Merge Sort efficiently sorts an array by recursively dividing it into halves, sorting each half, and merging them back together in sorted order. This approach ensures that the entire array is sorted progressively as smaller subarrays are merged.

```python
import random

#bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

#selection sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

#insertion sort        
def insertion_sort(arr):
    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

#merge sort        
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    # Divide the array into two halves
    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    
    # Recursively divide, sort and merge each half
    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)
    
    # Merge the sorted halves
    return merge(left_half, right_half)

def merge(left, right):
    sorted_arr = []
    left_index, right_index = 0, 0
    
    # Merge and sort lists into sorted_arr
    while left_index < len(left) and right_index < len(right):
        if left[left_index] <= right[right_index]:
            sorted_arr.append(left[left_index])
            left_index += 1
        else:
            sorted_arr.append(right[right_index])
            right_index += 1
    
    # Append remaining elements of left or right sublist if any
    sorted_arr.extend(left[left_index:])
    sorted_arr.extend(right[right_index:])
    
    return sorted_arr
```

```python
# Generate a list of 10000 random integers
arr = [random.randint(1, 10000) for i in range(10000)]

#arr1 = [random.randint(1, 10) for i in range(10)]

# Sort the list using each algorithm and time it
import time

start_time = time.time()
bubble_sort_result = bubble_sort(arr.copy())
bubble_sort_time = time.time() - start_time

start_time = time.time()
selection_sort(arr.copy())
selection_sort_time = time.time() - start_time

start_time = time.time()
insertion_sort(arr.copy())
insertion_sort_time = time.time() - start_time

start_time = time.time()
merge_sort(arr.copy())
merge_sort_time = time.time() - start_time

print("Bubble Sort time:", bubble_sort_time)
print("Selection Sort time:", selection_sort_time)
print("Insertion Sort time:", insertion_sort_time)
print("Merge Sort time:", merge_sort_time)
```

### Comparison

**Stability:**
- **Merge Sort**: Merge Sort is stable. It preserves the relative order of equal elements in the sorted array.
- **Bubble Sort**: Bubble Sort is stable.
- **Insertion Sort**: Insertion Sort is stable.
- **Selection Sort**: Selection Sort is not stable. It does not guarantee the relative order of equal elements in the sorted array.

**Performance:**
- **Merge Sort**: 
  - Time Complexity: \( O(n \log n) \) in all cases (average, worst, and best).
  - Space Complexity: \( O(n) \) due to the need for auxiliary space.
  - Performance is consistent and efficient for large datasets, making it suitable for most applications.

- **Bubble Sort**: 
  - Time Complexity: \( O(n^2) \) in the average and worst cases, \( O(n) \) in the best case (when the list is already sorted).
  - Space Complexity: \( O(1) \) (in-place sorting).
  - Inefficient for large datasets due to its quadratic time complexity. Best used for educational purposes or small datasets.

- **Insertion Sort**: 
  - Time Complexity: \( O(n^2) \) in the average and worst cases, \( O(n) \) in the best case (when the list is nearly sorted).
  - Space Complexity: \( O(1) \) (in-place sorting).
  - Efficient for small datasets or datasets that are nearly sorted. Performs well in practice for such cases.

- **Selection Sort**: 
  - Time Complexity: \( O(n^2) \) in all cases (average, worst, and best).
  - Space Complexity: \( O(1) \) (in-place sorting).
  - Generally less efficient than Insertion Sort for small datasets and not suitable for large datasets or nearly sorted datasets due to its consistent \( O(n^2) \) time complexity.

**Suitability:**
- **Merge Sort**: Ideal for situations requiring stable sorting with guaranteed \( O(n \log n) \) performance. Suitable for handling large datasets efficiently.
  
- **Bubble Sort**: Educational purposes or very small datasets where simplicity and ease of implementation are prioritized over performance.

- **Insertion Sort**: Well-suited for small datasets or datasets that are nearly sorted. Performs efficiently in practice for such scenarios.

- **Selection Sort**: Typically not recommended for practical use due to its \( O(n^2) \) time complexity, which makes it inefficient for most real-world applications compared to other \( O(n^2) \) algorithms like Insertion Sort.


### **In Python, the built-in sort() function for lists uses an algorithm called Timsort. Timsort is a hybrid sorting algorithm derived from merge sort and insertion sort, designed to perform well on many kinds of real-world data. Here are some key characteristics of Timsort:**

- **Hybrid Approach:** Timsort combines the strengths of merge sort and insertion sort. It uses merge sort for larger sub-arrays and insertion sort for smaller sub-arrays.

- **Adaptive:** Timsort is adaptive, meaning it takes advantage of existing order in the data. This makes it efficient for data that is already partially sorted or has runs of ascending or descending order.

- **Stable:** Timsort is a stable sorting algorithm, meaning that it preserves the relative order of equal elements in the sorted output.

- **Efficient:** It has a time complexity of O(n log n) in the average and worst cases, making it suitable for large datasets.

Used in Python: Python's **list.sort() and sorted()** functions internally use Timsort for sorting lists of comparable elements.

Timsort was designed by Tim Peters for Python in 2002 and has since become widely adopted due to its efficiency and versatility. It's particularly well-suited for Python's needs because of its stable nature and efficient handling of various data distributions encountered in real-world applications.
