To use `numpy` you need to import the module, using for example:

`NumPy` stores data in memory using **contiguous blocks** of memory, which allows for efficient numerical computation and manipulation. Here's a breakdown of how NumPy handles data in memory:

- Array Structure (ndarray)
    - NumPy arrays are stored as `ndarray` objects, which are the fundamental data structure for NumPy.
    - `The ndarray` is a fixed-size, multi-dimensional container for homogeneous data types (e.g., all elements of the array must be of the same type).
    - The data is stored as a contiguous block in memory, meaning that all elements are stored in one continuous sequence of memory locations.

##  `ndarray` versus `list`

So far the `numpy.ndarray` looks awefully much like a Python list (or nested list). Why not simply use Python lists for computations instead of creating a new array type? 


There are several reasons:

* Python lists are very general. They can contain any kind of object. They are dynamically typed. They do not support mathematical functions such as matrix and dot multiplications, etc. Implementing such functions for Python lists would not be very efficient because of the dynamic typing.
* Numpy arrays are **statically typed** and **homogeneous**. The type of the elements is determined when the array is created.
* Numpy arrays are memory efficient.
* Because of the static typing, fast implementation of mathematical functions such as multiplication and addition of `numpy` arrays can be implemented in a compiled language (C and Fortran is used).

### Index slicing

Index slicing is the technical name for the syntax `M[lower:upper:step]` to extract part of an array:

````python
A = np.array([1,2,3,4,5])
A
````

````python
A.shape
````

````python
A[1:3]
````

Array slices are *mutable*: if they are assigned a new value the original array from which the slice was extracted is modified:

````python
A[1:3] = [-2,-3]
A
````

We can omit any of the three parameters in `M[lower:upper:step]` (similar to `lists`):


````python
A[::] # lower, upper, step all take the default values
````

````python
A[::2] # step is 2, lower and upper defaults to the beginning and end of the array
````

````python
A[:3] # first three elements
````

````python
A[3:] # elements from index 3
````

Negative indices counts from the end of the array (positive index from the begining):

````python
A = np.array([1,2,3,4,5])
````

````python
A[-1] # the last element in the array
````

````python
A[-1:-4:-1] # the last three elements (reversed)
````

````python
A[-1:-4] # Step not assigned and not recognized
````

````python
A[-1:-4: 1] # Step not recognized
````

````python
A[1:4][-1]
````

````python
A[-3:] # the last three elements
````

````python
A[2::1] # the last three elements
````

Index slicing works exactly the same way for multidimensional arrays:

````python
A = np.array([[n+m*10 for n in range(5)] for m in range(5)])
A
````

````python
# a block from the original array
A[1:4, 1:4]
````

````python
# strides
A[::2, ::2]
````

We can also use index ***masks***: If the index mask is an Numpy array of data type `bool`, then an element is selected (True) or not (False) depending on the value of the index mask at the position of each element: 

````python
B = np.array([n for n in range(5)])
B
````

````python
row_mask = np.array([True, True, True, False, False])
B[row_mask]
````

````python
# same thing
row_mask = np.array([1,1,1,0,0], dtype=bool)
B[row_mask]
````

This feature is very useful to conditionally select elements from an array, using for example comparison operators:

````python
x = np.arange(0, 10, 0.5)
x
````

````python
# AND opeator
mask = (5 < x) & (x < 7.5) 
#mask = (5 < x) * (x < 7.5)

#OR operator
#mask = (5 < x) | (x < 7.5)
#mask = (5 < x) + (x < 7.5)

mask
````

````python
mask1 = (5 < x) and (x < 7.5) # do not use logical operator
````

````python
mask2 = []
for x_el in x:
    mask1 = (5 < x_el) and (x_el < 7.5)
    mask2.append(mask1)
    
mask2
````

````python
x[mask]
````

## Copy and "deep copy"

To achieve high performance, assignments in Python usually do not copy the underlaying objects. This is important for example when objects are passed between functions, to avoid an excessive amount of memory copying when it is not necessary (technical term: pass by reference). 

````python
A = np.array([[1,2], [3,4]])
A
````

````python
# now B is referring to the same array data as A 
B = A
````

````python
# changing B affects A
B[0,0] = 10
B
````

````python
A
````

If we want to avoid this behavior, so that when we get a new completely independent object `B` copied from `A`, then we need to do a so-called "deep copy" using the function `copy`:

````python
B = np.copy(A)
````

````python
# now, if we modify B, A is not affected
B[0,0] = -5
B
````

````python
A
````

## Iterating over array elements

Generally, we want to avoid iterating over the elements of arrays whenever we can (at all costs). The reason is that in a interpreted language like Python (or MATLAB), iterations are really slow compared to vectorized operations. 

However, sometimes iterations are unavoidable. For such cases, the Python `for` loop is the most convenient way to iterate over an array:

````python
v = np.array([1,2,5,7])
````

````python
for i in v:
    print(i)
````

````python
M = np.array([[1,2], [3,4]])

for row in M:
    print("row", row)
    for element in row:
        print(element)
````

When we need to iterate over each element of an array and modify its elements, it is convenient to use the `enumerate` function to obtain both the element and its index in the `for` loop: 

````python
for idx,v in enumerate(M):
    print(idx)
    print(v)
````

````python
for row_idx, row in enumerate(M):
    print("row_idx", row_idx, "row", row)
    
    for col_idx, element in enumerate(row):
        print("col_idx", col_idx, "element", element)
       
        # update the matrix M: square each element
        M[row_idx, col_idx] = element ** 2
````

````python
# each element in M is now squared
M
````

## Using arrays in conditions

When using arrays in conditions,for example `if` statements and other boolean expressions, one needs to use `any` or `all`, which requires that any or all elements in the array evalutes to `True`:

````python
M
````

````python
if (M > 5).any():
    print("at least one element in M is larger than 5")
else:
    print("no element in M is larger than 5")
````

````python
if (M > 5).all():
    print("all elements in M are larger than 5")
else:
    print("all elements in M are not larger than 5")
````

##### datetime64 and timedelta64

````python
import numpy as np

# Create datetime64 objects for two dates
start_date = np.datetime64('2024-06-01')
end_date = np.datetime64('2024-06-11')
````

````python
print(start_date,end_date)
````

````python
# Calculate the difference between two dates
duration = end_date - start_date
print(f"Duration: {duration}")
````

````python
# Add 10 days to the start date
new_date = start_date + np.timedelta64(32, 'D')
print(f"New Date: {new_date}")
````

````python
"""
There are two Timedelta units (‘Y’, years and ‘M’, months) 
which are treated specially, 
because how much time they represent changes depending on 
when they are used. While a timedelta day unit is equivalent to 24 hours,
there is no way to convert a month unit into days, 
because different months have different numbers of days.
"""
new_date_ = start_date + np.timedelta64(1, 'M')
print(f"New Date: {new_date_}")
````

````python
start_date_ = np.datetime64('2024-06')
new_date_ = start_date_ + np.timedelta64(1, 'M')
new_date_
````

````python
start_date_ = np.datetime64('2024-06-01', 'M')  # 'M' unit explicitly stated
new_date_ = start_date_ + np.timedelta64(1, 'M')
print(new_date_)
````

````python
new_date__ = start_date_ + np.timedelta64(32, 'D')
new_date__
````

````python
# Subtract 5 hours from the new date
adjusted_datetime = new_date - np.timedelta64(5, 'h')
print(f"Adjusted DateTime: {adjusted_datetime}")
````

````python
# Difference in days and hours between two datetime64 objects
start_datetime = np.datetime64('2024-06-11T08:00')
end_datetime = np.datetime64('2024-06-11T20:00')
difference_hours = end_datetime - start_datetime
print(f"Difference in hours: {difference_hours}")
````
